```python
# fit a model and plot learning curve
def tuning_epochs(X_train, y_train, X_test, y_test, n_epoch):
    # define model
    model = Sequential()
    model.add(Dense(100, input_dim=12, activation='relu', kernel_initializer='normal'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1, activation='relu'))

    # compile model
    opt = SGD(lr=0.0001, momentum=0.9)
    model.compile(loss='mse', optimizer=opt)
    
    # fit model
    history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=n_epoch, verbose=0, batch_size=len(X_train))
    
    # plot learning curves
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='test')
    plt.title('epoch='+str(n_epoch))
    plt.legend()
```

```python
# fit a model and plot learning curve
def tuning_decay(X_train, y_train, X_test, y_test, decay):
    # define model
    model = Sequential()
    model.add(Dense(100, input_dim=12, activation='relu', kernel_initializer='normal'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1, activation='relu'))

    # compile model
    opt = SGD(lr=0.001, momentum=0.8,decay=decay)
    model.compile(loss='mse', optimizer=opt)
    
    # fit model
    history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=200, verbose=0, batch_size=64)
    
    # plot learning curves
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='test')
    plt.title('decay='+str(decay))
    plt.legend()
```

```python
def train_model(df,nama_model,scaler):
    y=df["price"]
    X=df.drop(["price"],axis=1)
    if (nama_model == "RGR"):
        arr_skor=[]
        arr_depth=[i for i in range(1,21,1)]
        for depth in arr_depth:
            regr = RandomForestRegressor(max_depth=depth, random_state=1)
            cv = KFold(n_splits=10, random_state=1, shuffle=True)
            scores = cross_val_score(regr, X, y, scoring='neg_mean_squared_error', cv=cv)
            arr_skor.append(np.mean(scores*(-1)))
        for i in range(len(arr_skor)):
            if arr_skor[i] == np.min(arr_skor):
                spesific_RFR = arr_depth[i]
        skor_RFR=np.min(arr_skor)
        regr=RandomForestRegressor(max_depth=spesific_RFR, random_state=1).fit(X,y)
        return (skor_RFR, regr)
    
    if (nama_model == "ETR"):
        arr_skor=[]
        arr_estimator=[i for i in range(50,150,10)]
        for estim in arr_estimator:
            regr = ExtraTreesRegressor(n_estimators=estim, random_state=1)
            cv = KFold(n_splits=10, random_state=1, shuffle=True)
            scores = cross_val_score(regr, X, y, scoring='neg_mean_squared_error', cv=cv)
            arr_skor.append(np.mean(scores*(-1)))
        for i in range(len(arr_skor)):
            if arr_skor[i] == np.min(arr_skor):
                spesific_ETR = arr_estimator[i]
        skor_ETR = np.min(arr_skor)
        regr = ExtraTreesRegressor(n_estimators=spesific_ETR, random_state=1).fit(X,y)
        return (skor_ETR,regr)
    
    if (nama_model == "Ridge"):
        model = Ridge(alpha=1)
        cv = KFold(n_splits=10, random_state=1, shuffle=True)
        scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=cv)
        skor_ridge=np.mean(scores*(-1))
        return (skor_ridge,model.fit(X,y))
    
    if (nama_model == "Linear"):
        model = linear_model.LinearRegression()
        cv = KFold(n_splits=10, random_state=1, shuffle=True)
        scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=cv)
        skor_linear=np.mean(scores*(-1))
        return (skor_linear,model.fit(X,y))
    
    if (nama_model == "SVR"):
        model = SVR()
        cv = KFold(n_splits=10, random_state=1, shuffle=True)
        scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=cv)
        skor_SVR=np.mean(scores*(-1))
        return (skor_SVR,model.fit(X,y))

def generate_second_training(model,scaler):
    training2 = fix_train.drop(["author_id","description"],axis=1)
    for k in range(training2.shape[0]):
        if pd.isnull(training2["price"][k]):
            val=list(training2.iloc[k,1:].values)
            key=list(training2.iloc[k,1:].index)
            kamus={}
            for i,j in zip(key,val):
                kamus[i]=j
            data=pd.DataFrame(kamus,index=[0])
            predicted_price=model.predict(data)
            training2["price"][k] = predicted_price
    skala_harga=scaler.fit_transform(scaler.inverse_transform(training2["price"]).reshape(3400,1))
    training2["price"]=skala_harga
    return (training2, scaler)
```

```python
arr_model=["RGR","ETR","Ridge","Linear","SVR"]
hasil_first_training=[]
hasil_second_training=[]
for model_1 in arr_model:
    df_1 = first_train
    #scaling
    scaler = StandardScaler()
    scaled_price = scaler.fit_transform(valid_price)
    #training 1
    skor_1, hasil_model_1 = train_model(df_1,model_1,scaler)
    hasil_first_training.append("{} : {}".format(model_1,skor_1))
    #re-fitting
    df_2,scaler = generate_second_training(hasil_model_1,scaler)
    for model_2 in arr_model:
        skor_2, hasil_model_2 = train_model(df_2,model_2,scaler)
        hasil_second_training.append("{} : {}".format(model_2,skor_2))
        print("{} : {} ----- {} : {}".format(model_1,skor_1,model_2,skor_2))
```
